```python
df=spark.table("workspace.ventas.retail_transactions_course_dataset")
```

```python
display(df)
```

```python
df_clean = df
```

```python
# Limpiamos la ase de datos de duplicados
df_clean = df_clean.dropDuplicates(["transaction_id"])
```

```python
df_clean = df_clean.dropna(subset=["transaction_id", "transaction_ts", "total_eur"])
```

```python
display(df_clean)
```

```python
df.count()
df_clean.count()

original_count = df.count()
clean_count = df_clean.count()

print("Row before cleaning", original_count)
print("Row after cleaning", clean_count)
print("Row removed", original_count - clean_count)
```

```python
df_clean.write.format("delta").mode("overwrite").saveAsTable("retail_delta")
```

```python
%sql
UPDATE retail_delta
SET discount_pct ='0'
WHERE discount_pct ='N/A';
```
